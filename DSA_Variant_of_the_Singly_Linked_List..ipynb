```python
class Node:
    def __init__ (self, data):
        self.data = data
        self.next = None
        self.down = None
```

```python
class TwoDim_List:
    def __init__(self):
        self.head = None

    
    def insertAtLast(self, data):
        curr = self.head
        node = Node(data)

        if curr != None:
            while curr.down and curr.down.data // 100 < data // 100:
                curr = curr.down

            if curr.data // 100 != data // 100:
                if curr.down == None:
                    curr.down = node
                else:
                    if curr.down.data // 100 > data // 100:
                        node.down = curr.down
                        curr.down = node
                    else:
                        curr = curr.down
                        while curr.next != None:
                            curr = curr.next
                        curr.next = node
            else:
                while curr.next != None:
                    curr = curr.next
                curr.next = node            
        else:
            self.head = node

    def insertAtFirst(self, data):
        curr = self.head
        prevFirst = None
        node = Node(data)

        if curr != None:
            while curr.down and curr.down.data // 100 < data // 100:
                prevFirst = curr
                curr = curr.down
                
            if curr.data // 100 != data // 100:
                if curr.down == None:
                if curr.down == None:
                    curr.down = node
                else:
                    if curr.down.data // 100 > data // 100:
                            node.down = curr.down
                            curr.down = node
                    else:
                        downNode = curr.down
                        node.next = downNode
                        node.down = downNode.down
                        curr.down = node
                        downNode.down = None
            else:
                if prevFirst:
                    prevFirst.down = node
                else:
                    self.head = node
                node.next = curr
                node.down = curr.down
                curr.down = None
                    
        else:
            self.head = node

    def insertAtNth(self, data, n):
        node = Node(data)
        if n == 1:
            self.insertAtFirst(node)
        elif n == self.length + 1:
            self.insertAtLast(node)
        elif n <= self.length:
            curr = self.head
            i = 1
            while i < n - 1:
                curr = curr.next
                i += 1
                
            node.next = curr.next
            curr.next = node

    def deleteFirst(self):
        self.head = self.head.next
    
    def deleteLast(self):
        curr = self.head
        prev = None
        while curr.next != None:
            prev = curr
            curr = curr.next
        
        prev.next = None
        

    def deleteAtN(self, n):
        if n == 1:
            self.deleteFirst()
        elif n == self.length:
            self.deleteLast()
        elif n < self.length:
            i = 1
            prev = None
            curr = self.head
            while i < n:
                prev = curr
                curr = curr.next
                i += 1
            prev.next = curr.next
            

    def traverse(self):
        curr = self.head
        first = self.head
        while first != None:
            while curr != None:
                print(curr.data, end = " -> ")
                curr = curr.next
            
            first = first.down
            curr = first
            print("\n|")
```
